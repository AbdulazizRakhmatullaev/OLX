```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# so the difference is in the code at the end

# Apple -      https://www.olx.uz/elektronika/telefony/mobilnye-telefony/?currency=UZS&search%5Bfilter_enum_mobile_phone_manufacturer%5D%5B0%5D=2065 <--
# Blackberry - https://www.olx.uz/elektronika/telefony/mobilnye-telefony/?currency=UZS&search%5Bfilter_enum_mobile_phone_manufacturer%5D%5B0%5D=2069 <--
# Samsung -    https://www.olx.uz/elektronika/telefony/mobilnye-telefony/?currency=UZS&search%5Bfilter_enum_mobile_phone_manufacturer%5D%5B0%5D=2101 <--
```

```python
filter = {"Apple": 2065, "Blackberry": 2069, "Samsung": 2101}
products = {"brand": [], "name": [], "price": []}

for product, code in filter.items():
    r = requests.get(
        f"https://www.olx.uz/elektronika/telefony/mobilnye-telefony/?currency=UZS&search%5Bfilter_enum_mobile_phone_manufacturer%5D%5B0%5D={code}"
    ).text
    soup = BeautifulSoup(r, "html.parser")
    names = soup.find_all("h6", class_="css-16v5mdi er34gjf0")
    prices = soup.find_all("p", class_="css-10b0gli er34gjf0")

    for name, price in zip(names, prices):
        raw_price = price.text

        if any(char.isalpha() for char in raw_price):
            if raw_price.lower() == 'обмен':
                continue
            else:
                numeric_price = "".join(char for char in raw_price if char.isnumeric())

        # print(raw_price)
        # print(numeric_price)
        products["brand"].append(product)
        products["name"].append(name.text)
        products["price"].append(int(numeric_price))
```

```python
df = pd.DataFrame(products)
df
```
